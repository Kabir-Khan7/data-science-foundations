# Numpy vs Python Lists - Performance Test

```python
import numpy as np 
import time
```

## Python Zip Example

```python
# Python Zip Explained
l1 = [1, 2, 4]
l2 = [6, 7, 8]
list(zip(l1,l2))
```

## Using Python List evaluating speed

```python
# Using Python List
size = 1_000_0000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()

add = [x+y for x,y in zip(l1, l2)]
end = time.time()
print(add[0:10])

print(f"Time taken: {end - start:.2f} seconds")
```

## Using Numpy Array evaluating speed

# Using Numpy Array
size = 1_000_0000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()

add = l1 + l2
print(add[0:10])
end = time.time()

print(f"Time taken: {end - start:.2f} seconds")
