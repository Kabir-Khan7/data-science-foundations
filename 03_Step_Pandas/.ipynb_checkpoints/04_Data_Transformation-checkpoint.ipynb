```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
df3 = pd.DataFrame(data)
df3
```

```python
melt_form = df3.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
melt_form
```

```python
pivot_form = melt_form.pivot(index="Name", columns="Subject", values="Score")
pivot_form
```

## Exaple of pivot_table() to handle duplicates:
Suppose the DataFrame is having two Alice or Bob, with this sense we will using below addiotional (aggfunc="mean") to take mean for duplicate entries. 

For Example: 
If Alice give twice Math's paper, Paper1 = 80 and Paper2=85 then mean value will be 82.5 mark in just one column of Math represent both data point mean

```python
import pandas as pd

df4 = pd.DataFrame(
    [
        ["Alice", 85, 80, 90, 88],      
        ["Bob",   78, 78, 82, 85],
        ["Charlie", 92, 92, 89, 94]
    ],
    columns=["Name", "Math", "Math", "Science", "English"]  
)

print(df4)
```

```python
melt_form1 = df4.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
melt_form1
```

```python
pivot_form2 = melt_form1.pivot_table(index="Name", columns="Subject", values="Score", aggfunc="mean")
pivot_form2
```

# Finance Example : Classic Example

```python
import pandas as pd
```

```python
pd.__version__
```

```python
df = pd.read_excel("fin_data.xlsx")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
for col in ['Close', 'P/E Ratio', 'Dividend Yield', 'Dividend']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Dividend'] = df['Dividend'].fillna(0).abs()
df['Dividend Yield'] = df['Dividend Yield'].abs()

df[['High', 'Low', 'Close', 'Volume']] = df.groupby('Ticker')[['High', 'Low', 'Close', 'Volume']].ffill() #Fill missing High/Low/Close/Volume within each stock by carrying forward the last known price.
df['P/E Ratio'] = df.groupby('Ticker')['P/E Ratio'].transform(lambda x: x.fillna(x.mean())) #Fill missing P/E values for a stock using the stocks mean P/E.
df['Dividend Yield'] = df.groupby('Ticker')['Dividend Yield'].transform(lambda x: x.fillna(x.mean())) #Fill missing Dividend Yield values for a stock using the stocks mean Dividend Yield.

print(df.isnull().sum())
```

```python
df
```

In above example we explore the financial dataset to do some initial data cleaning and preparotory steps before proceeding to use the data further.
