```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
df3 = pd.DataFrame(data)
df3
```

```python
melt_form = df3.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
melt_form
```

```python
pivot_form = melt_form.pivot(index="Name", columns="Subject", values="Score")
pivot_form
```

## Exaple of pivot_table() to handle duplicates:
Suppose the DataFrame is having two Alice or Bob, with this sense we will using below addiotional (aggfunc="mean") to take mean for duplicate entries. 

For Example: 
If Alice give twice Math's paper, Paper1 = 80 and Paper2=85 then mean value will be 82.5 mark in just one column of Math represent both data point mean

```python
import pandas as pd

df4 = pd.DataFrame(
    [
        ["Alice", 85, 80, 90, 88],      
        ["Bob",   78, 78, 82, 85],
        ["Charlie", 92, 92, 89, 94]
    ],
    columns=["Name", "Math", "Math", "Science", "English"]  
)

print(df4)
```

```python
melt_form1 = df4.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
melt_form1
```

```python
pivot_form2 = melt_form1.pivot_table(index="Name", columns="Subject", values="Score", aggfunc="mean")
pivot_form2
```
