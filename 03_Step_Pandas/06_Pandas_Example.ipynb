```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
team1_matches = df['Team 1'].value_counts()
team1_matches
```

```python
team2_matches = df['Team 2'].value_counts()
team2_matches
```

```python
total_matches = team1_matches.add(team2_matches, fill_value=0)
total_matches.sort_values(ascending=False)
```

```python
df[['Team1_Goals', 'Team2_Goals']] = df['FT'].str.split('-', expand=True).astype(int)
```

```python
df
```

```python
df["Total_Goals"] = df["Team1_Goals"] + df["Team2_Goals"]
```

```python
df
```

### Highest Scoring Matches (Top 5)

```python
df.sort_values(by="Total_Goals", ascending=False).head(5)
```

### Team 1 vs Team 2 Win Analysis

```python
df["Result"] = df.apply(
    lambda x: "Team 1 Win" if x["Team1_Goals"] > x["Team2_Goals"]
    else "Team 2 Win" if x["Team1_Goals"] < x["Team2_Goals"]
    else "Draw",
    axis=1
)
```

```python
df["Result"].value_counts()
```

### Average Goals by Team 1 (Home Team)

```python
df.groupby("Team 1")["Team1_Goals"].mean().sort_values(ascending=False)
```

### Average Goals by Team 2 (Away Team)

```python
df.groupby("Team 2")["Team2_Goals"].mean().sort_values(ascending=False)
```

## Manchester United Team Analysis

```python
manu = df[(df["Team 1"] == "Manchester United FC") | (df["Team 2"] == "Manchester United FC")].copy()
```

```python
manu
```

```python
manu[manu["Total_Goals"] >= 3]
```

```python
manu[["Home_Goals", "Away_Goals"]] = df["FT"].str.split('-', expand=True).astype(int)
```

```python
manu
```

### Home Game Analysis

```python
goals = manu["Home_Goals"].sum()
avg = manu["Home_Goals"].mean()
max_goal = manu["Home_Goals"].max()
std_goal = manu["Home_Goals"].std()

print(f"Total Home Goals: {goals}")
print(f"Avg Home Goals Per Game: {avg}")
print(f"Max Goals at Home: {max_goal}")
print(f"Std Goals at Home: {std_goal}")
```

```python
goals = manu["Away_Goals"].sum()
avg = manu["Away_Goals"].mean()
max_goal = manu["Away_Goals"].max()
std_goal = manu["Away_Goals"].std()

print(f"Total Away Goals: {goals}")
print(f"Avg Away Goals Per Game: {avg}")
print(f"Max Goals at Away: {max_goal}")
print(f"Std Goals at Away: {std_goal}")
```

```python
total_goals = manu["Home_Goals"].sum() + manu["Away_Goals"].sum()
print("Total Scored by Manchester United:",total_goals)
```
