```python
companies = pd.DataFrame({
    "Company": ["Alpha", "Beta", "Gamma"],
    "Revenue": [800000, 600000, 900000],
    "Net_Income": [120000, 45000, 150000],
    "Total_Assets": [900000, 700000, 1000000],
    "Equity": [400000, 250000, 500000]
})
```

```python
companies
```

```python
companies["ROA"] = (companies["Net_Income"] / companies["Total_Assets"])
```

```python
companies
```

```python
companies["ROA_Rank"] = companies["ROA"].rank(ascending=False)
```

```python
companies
```

```python
companies_sorted = companies.sort_values(by="ROA", ascending=False)
companies_sorted
```

```python
companies["ROE"] = (companies["Net_Income"] / companies["Equity"])
```

```python
companies
```

```python
companies["ROE_Rank"] = companies["ROE"].rank(ascending=False)
```

```python
companies
```

```python
companies["Leverage_Effect"] = (companies["ROE"] / companies["ROA"])
```
