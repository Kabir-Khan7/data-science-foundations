# When you open a file:
file_object = open("filename.txt", "r")
### 1. Python requests OS to open file
### 2. OS checks permissions and availability
### 3. OS returns a file descriptor
### 4. Python wraps it in a file object
### 5. File pointer is set to beginning (mode-dependent)

### Method 1: (Traditional Approach)

```python
file = None

try: 
    file = open("main.txt", "r")
    content = file.read()
    print(content)
finally:
    if file:
        file.close()
```

### Method 2: Context Manager

```python
with open("main.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("main.txt", "r") as f:
    for line in f: 
        print(line.strip())
```

```python
with open("main.txt", "r") as file:
    content = file.readlines()
    print(content)
```
